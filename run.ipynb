```python
import pandas as pd
import ast

# Load your current CSV
df = pd.read_csv("transactions.csv", header=None)

# Convert string list → actual list → clean string
df[0] = df[0].apply(lambda x: ,",",.join(ast.literal_eval(x)))

# Save clean version
df.to_csv("transactions_clean.csv", index=False, header=False)

print("✅ Fixed! Use transactions_clean.csv")
```

```python
import pandas as pd
import ast

# Load CSV
df = pd.read_csv("transactions.csv", header=None)

cleaned_data = []

for row in df[0]:
    try:
        # Try to convert string → list
        items = ast.literal_eval(row)
        
        # Join into proper CSV format
        cleaned_row = ",".join([str(i).strip() for i in items])
    
    except:
        # If error, clean manually
        cleaned_row = row.replace("[", "").replace("]", "").replace("'", "").replace('"', "")
    
    cleaned_data.append(cleaned_row)

# Save cleaned file
pd.DataFrame(cleaned_data).to_csv("transactions_clean.csv", index=False, header=False)

print("✅ SUCCESS: transactions_clean.csv created")
```

```python
import pandas as pd
import ast

# Load CSV
df = pd.read_csv("transactions.csv", header=None)

cleaned_data = []

for row in df[0]:
    try:
        # Remove outer quotes if present
        row = row.strip().strip('"')

        # Convert string → list
        items = ast.literal_eval(row)

        # Convert list → comma-separated string
        cleaned_row = ",".join([str(i).strip() for i in items])

    except:
        # Fallback cleaning (force clean)
        cleaned_row = row.replace("[", "").replace("]", "").replace("'", "").replace('"', "")

    cleaned_data.append(cleaned_row)

# Save clean dataset
pd.DataFrame(cleaned_data).to_csv("transactions_clean.csv", index=False, header=False)

print("✅ DONE: transactions_clean.csv created successfully")
```

```python
import pandas as pd

# Load cleaned file
df = pd.read_csv("transactions_clean.csv", header=None)

# Remove unwanted double quotes
df[0] = df[0].str.replace('"', '')

# Save final clean file
df.to_csv("transactions_final.csv", index=False, header=False)

print("✅ FINAL CLEAN FILE READY: transactions_final.csv")
```
